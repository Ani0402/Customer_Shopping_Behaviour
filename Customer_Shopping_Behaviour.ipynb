```python
import pandas as pd

df = pd.read_csv("C:\\Users\\cubic\\Documents\\DataAnalystProjects\\New folder\\customer_shopping_behavior (1).csv")
df.head()
```

```python
df.describe(include = 'all')
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.columns
```

```python
# creating a age group column 
labels = ['Young-Adult', 'Adult', 'Middle-Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)
df[['age', 'age_group']].head(10)
```

```python
#creating column purchase frequency days
purchase_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Bi-Weekly' : 14,
    'Quarterly' : 90,
    'Annually' : 365,
    'Every 3 months' : 90,
}
df['purchase_freq_days'] = df['frequencyofpurchases'].map(purchase_mapping)
```

```python
df[['frequencyofpurchases', 'purchase_freq_days']].head(10)
```

```python
df = df.drop('promocodeused', axis = 1)
df
```
